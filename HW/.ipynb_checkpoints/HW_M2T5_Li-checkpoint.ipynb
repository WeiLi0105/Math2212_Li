```python
import  pandas  as pd
import  numpy  as np
import  matplotlib.pyplot  as plt
import seaborn as sns
```

```python
loan= pd.read_csv('loans_full.csv',
                  na_values = ':' )
loan
```

# Question  1 

Draw a correlation heatmap for the loans dataframe

```python
loan_corr_matrix = loan.corr()

print loan_corr_matrix
```

```python
sns.heatmap(loan_corr_matrix)
```

# Question 2 

Compute the following correlations

A. Annual Income vs. Debt-to-Income ratio 

```python
income_corr_ratio = loan["annual_income"].corr(loan["debt_to_income"])

print income_corr_ratio
```

B. Annual Income vs Interest Rate

```python
income_corr_interest = loan["annual_income"].corr(loan["interest_rate"])

print income_corr_interest
```

```python
ratio_corr_interest = loan["debt_to_income"].corr(loan["interest_rate"])

print ratio_corr_interest
```

# Question 3

Make a scatterplot using seaborns library for each of the pairs in Question 2 with the points colored according to 'Grade'

```python
sns.scatterplot(data=loan, x='annual_income',y='debt_to_income',hue="grade")

plt.show()
```

```python
sns.scatterplot(data=loan, x='annual_income',y='interest_rate',hue="grade")

plt.show()
```
