# Question 1

```python
# Define a piecewise function using an "if, elif, else" statement

def f(x):
    if(x>0):
        return x**2
    elif(x==0):
        return 0
    else:
        return 2*x

print f(-1)
print f(2)
print f(0)
```

# Question 2

What are the different ways of handling ties using "rank()"?

Syntax for "rank()" :
DataFrame.rank(axis=0, method=’average’, numeric_only=None, na_option=’keep’, ascending=True, pct=False)


The ways of handling ties using "rank()"('average', 'min', 'max', 'first', 'dense'):
1. average: average rank of group
2. min: lowest rank in group
3. max: highest rank in group
4. first: ranks assigned in order they appear in the array
5. dense: like 'min', but rank always increases by 1 between groups

# Question 3

Without changing the original census dataframe, create a copy of the census dataframe and remove all rows with missing data and adds the following columns

```python
census2 = census
new_census = census2.dropna()
new_census
```

```python
#add square root of total family income
def f(x):
    return math.sqrt(x)

census_sqrt=new_census
census_sqrt['square_root_income']  = new_census['total_family_income'].apply(f)
census_sqrt[[ 'total_family_income','square_root_income']]
```

```python
#total family income adjusted to 2020-dollars, use average inflation rate of 1.2% increase each year (use lambda functions)

inflation_income = lambda x: x*(1.012**20)

census_inflation=new_census
census_inflation['inflation_dollar_income'] = new_census['total_family_income'].apply(inflation_income)
census_inflation[[ 'total_family_income','inflation_dollar_income']]
```

```python
#birth year of person filling out the census

census_birth_year=new_census
census_birth_year['birth_year'] = new_census['census_year']-new_census['age']
census_birth_year[['birth_year']]
```

# Question 4

```python
#Only males with the "total family income" in increasing order.  

census_male = census[(census['sex']=='Male')]
census_by_income=census_male.sort_values(by = 'total_family_income', ascending = True, inplace =False)#inplace -- overwrite
census_by_income
```

```python
#Only females with the "total family income" in decreasing order. 

census_female = census[(census['sex']=='Female')]
census_by_income=census_female.sort_values(by = 'total_family_income', ascending = False, inplace =False)
census_by_income
```

# Question 5

```python
# Create and print a new dataframe that has the average "total family income" in each state. 

census_hw5 = census[['state_fips_code', 'total_family_income']].groupby('state_fips_code').mean()
census_hw5
```

# Question 6

```python
# Create and print a new dataframe with the average age for each race.

census_hw6 = census[['race_general', 'age']].groupby('race_general').mean()
census_hw6
```

# Question 7

```python
#Print a pivot table that gives the average personal income in each state, separating by sex. 

hwTable = pd.pivot_table(
    census,
    values = 'total_personal_income',
    index = 'state_fips_code',
    columns ='sex',
    aggfunc = np.mean
)

hwTable
```
