# Step 11: Pairwise correlation between the different columns of the data

Correlation is a value between -1 and 1 that represents how closely two features move in unison:

Positive correlation means that as one feature increases, the other increases. 

Negative correlation means that as one feature increases, the other decreases.

Correlations near -1 or 1 indicate a strong relationship.

Those closer to 0 indicate a weak relationship.

0 indicates no relationship.

```python
sns.heatmap(censusData.corr(),annot=True,cmap='RdYlGn')
```

As we can observe from the above heatmap of correlations that, there is a high correlation between

total family income - total personal income: 0.77

 Now,using Seaborn, we visualize the relation between total family income and total personal income by using a regression plot.


```python
sns.regplot(x='total_family_income',y='total_personal_income',data=censusData)
```

Now,using Seaborn, we visualize the relation between age and  total personal income by using a regression plot.

```python
sns.regplot(x='age',y='total_personal_income',data=censusData)
```

# Step 13: Exploring for Categorical Data

```python
#Group by state for the mean of total family income

census_state = censusData[['state_fips_code','total_family_income']].groupby('state_fips_code').mean()
census_state.plot(kind = 'bar')
```

From the graph, we can see that the average amount of family income of different states are different (The state of Califonia has the highest mean value of total family income). 

```python
#Group by sex for the mean of total personal income

census_state = censusData[['sex','total_personal_income']].groupby('sex').mean()
census_state.plot(kind = 'bar')
```

From the graph, we can see that the average amount of personal income of male is higher than female

```python
#Group by race for the mean of total personal income

census_race = censusData[['race_general','total_personal_income']].groupby('race_general').mean()
census_race.plot(kind = 'barh')
```

From the graph, we can see that the average amount of personal income for different races are different 

```python
#Group by marital status for the mean of total personal income

census_marital = censusData[['marital_status','total_personal_income']].groupby('marital_status').mean()
census_marital.plot(kind = 'barh')
```

From the graph, we can see that the average amount of personal income of different marital status are different.(The married / spouse present has the largest average personal income)
