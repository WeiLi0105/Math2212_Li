# Adding Rows to a datafram

The idea is to make the collection of new row into a new dataframe.

```python
# to create a new dataframe, start with dictionary

new_rows_dict = {
    'TIME': [2000, 2000, 2000],
    'GEO': ['USA', 'CANADA', 'Uganda'],
    'Value': [6.00, 5.78, 8.92]
}
new_rows_df = pd.DataFrame(new_rows_dict)
new_rows_df
```

```python
# Append the new_rows_df to the end of edu

edu_new_rows = edu.append(new_rows_df,ignore_index = True, sort = False)
edu_new_rows.tail()
```

# Aggregating Data

how can we sum, average, or ohter aggregating data that shares a similar property? 

**standard aggregating function from numpy**

cout() --this function counts how many values there are

sum()

mean()

median()
min()
max()

prod()--mulitiplies all the values

std()--standard deviation

var()--variance of all the values

```python
group_mean = edu[['GEO', 'Value']].groupby('GEO').mean()
print group_mean
```

```python
group_max = edu[['TIME','Value']].groupby('TIME').sum()
print group_max
```

```python
# Worksheet  6 -- run all the following, see what happened

edu_ws6 = edu[['TIME','GEO']].groupby('TIME').mean()

edu_ws6
```

```python
#worksheet 5--print a table that contains the product of all 'Values' within each year

edu_ws7 = edu[['TIME','Value']].groupby('TIME').prod()
edu_ws7
```

# Pivot Tables

similar to .groupby() except that instead of aggregating over one variable we aggregate over two variables

```python
ptable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'TIME',
    columns = 'GEO',
    aggfunc = np.sum
)

ptable
```

```python
#Worksheet 6: 

myTable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'GEO',
    columns ='TIME',
    aggfunc = np.mean
)

myTable
```

# Plotting Data

basic format:

dataframe.plot(x= 'Column1', y= 'Column2', kind =Graphtype, tile = 'Title')

```python
#scatter plot

edu.plot(x = 'TIME', y = 'Value', kind = 'scatter', title = 'Education')
```

```python
#line plot

edu_austria = edu[edu['GEO'] == 'Austria']

edu_austria.plot(x = 'TIME',y = 'Value', kind = 'line', title = 'eduAustria')
```

```python
#bar graphs

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'bar')
```

```python
#bar graphs horizontal

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'barh')
```
